### Tips seaborn dataset EDA & Vizualization

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = sns.load_dataset('tips')
df1.head(5)
```

```python
df1.tail(5)
```

```python
# total no of unique values in each column
df1.nunique()
```

### Univariate Analysis

```python
df1.columns
```

**Total_Bill**

```python
sns.distplot(df1['total_bill'],kde=True,hist=True)
plt.axvline(df1['total_bill'].mean(),color='red')
plt.axvline(df1['total_bill'].median(),color='green')
plt.show()
```

```python
sns.boxplot(df1['total_bill'], color='r')
plt.show()
```

```python
sns.violinplot(df1['total_bill'], color='g')
plt.show()
```

**Tip**

```python
sns.distplot(df1['tip'],kde=True,hist=True)
plt.axvline(df1['tip'].mean(),color='red')
plt.axvline(df1['tip'].median(),color='green')
plt.show()
```

```python
sns.boxplot(df1['tip'], palette='spring')
plt.show()
```

```python
sns.violinplot(df1['tip'], palette='autumn')
plt.show()
```

**Sex**

```python
df1['sex'].value_counts()
```

```python
sns.countplot(df1['sex'], palette='summer')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.title('Count Plot')
plt.show()
plt.show()
```

```python
plt.pie(x=df1['sex'].value_counts().values, labels=df1['sex'].value_counts().index, autopct="%2.2f%%",explode=[0,0])
plt.show()
```

**Smoker**

```python
f,ax = plt.subplots(1,2,figsize=(8,5))
df1['smoker'].value_counts().plot(kind='pie',ax=ax[0],autopct="%2.2f%%",explode=[0,0])
sns.countplot(df1['smoker'],ax=ax[1], palette='autumn')
plt.show()
```

**Day**

```python
f,ax = plt.subplots(1,2,figsize=(8,5))
df1['day'].value_counts().plot(kind='pie',ax=ax[0],autopct="%2.2f%%",explode=[0,0,0,0.2])
sns.countplot(df1['day'],ax=ax[1])
plt.show()
```

**Time**

```python
f,ax = plt.subplots(1,2,figsize=(8,5))
df1['time'].value_counts().plot(kind='pie',ax=ax[0],autopct="%2.2f%%",explode=[0,0])
sns.countplot(df1['time'],ax=ax[1], palette='autumn')
plt.show()
```

**Size**

```python
f,ax = plt.subplots(1,2,figsize=(14,6))
df1['size'].value_counts().plot(kind='pie',ax=ax[0],autopct="%2.2f%%",explode=[0,0,0,0.1,0.2,0.3])
sns.countplot(df1['size'],ax=ax[1])
plt.show()
```

### Multivariate

**Total_Bill & Tip with all other features**

```python
# Using sex in hue
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='sex', col='sex')
plt.show()
```

```python
# Using smoker in hue
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='smoker', col='smoker')
plt.show()
```

```python
# Using day as hue
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='day', col='day')
plt.show()
```

```python
# Using time as hue
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='time', col='time')
plt.show()
```

```python
# Using size as hue 
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False, hue='size')
plt.show()
```

```python
# Using size as hue 
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False, hue='size', col='size')
plt.show()
```

```python
# Using sex & smoker both with Total_bill and Tip
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,col='sex',row='smoker')
plt.show()
```

```python
# Using Day, sex, smoker with total_bill and tip
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='day',col='sex',row='smoker')
plt.show()
```

```python
# Using time, sex, smoker with total_bill and tip
sns.lmplot(x='total_bill',y='tip',data=df1,fit_reg=False,hue='time',col='sex',row='smoker')
plt.show()
```
